```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option("display.max_columns",None)

df=pd.read_csv("employee.csv")
```

```python
df.head()
```

```python
df.Attrition.value_counts().plot(kind="pie")
```

```python
#encodingto convert from labels to numbers
df["Attrition_transform"]=df.Attrition.replace({"Current employee":0,"Voluntary Resignation":1,
                   "Termination":2 })
df["Attrition_transform"]
```

```python
df.Attrition.value_counts()
```

```python
df.MaritalStatus.value_counts()
```

```python
#remove undesired valued
df.drop(df[df.MaritalStatus=="4"].index,inplace=True)
```

```python
df.drop(columns="Over18",axis=1,inplace=True)
```

```python
df.head()
```

```python
df.reset_index()   #optional
```

```python
df.Age.describe()
#课上讲了std>mean 说明很分散
```

```python
df.groupby("Attrition").Age.describe()
```

```python
#dumify the data
pd.get_dummies(df.MaritalStatus,prefix="MStatus_")
```

```python
pd.get_dummies(df,columns=["MaritalStatus"],prefix="MStatus_")
```

```python
#categories different age group, pd.cut()
pd.cut(x=df.Age,bins=[0,30,40,50,60],labels=["young people","30s","40s","50s"])
```

```python
df["Age_transform"]=pd.cut(x=df.Age,bins=[0,30,40,50,60],labels=["young people","30s","40s","50s"])
```

```python
df.head()
```

```python
#crosstab
pd.crosstab(df.Attrition,df.MaritalStatus)
```

```python
pd.crosstab(df.Attrition,df.MaritalStatus).plot.bar()
```

```python
#crosstab mutilple levels
pd.crosstab(df.Attrition,[df.MaritalStatus,df.Department])
```

```python
#pivot table
pd.pivot_table(df,index=["Attrition"],columns=["Department"],values=["Age"])
#default is mean
```

```python
pd.pivot_table(df,index=["Attrition"],columns=["Department"],values=["Age"],aggfunc=["median","mean"])
```

```python
df.pivot_table("Age",index=["Attrition"],columns=["Department"]).plot()
```

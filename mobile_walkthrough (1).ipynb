```python
# importing pandas and pyplot libraries.
import pandas as pd
import matplotlib.pyplot as plt

# the following line is a magic line to allow visualizations to be shown below codes after running it.
%matplotlib inline

df = pd.read_csv("mobile.csv") # read in the mobile data from csv file into dataframe
df.describe() # first look at some basic statistics of the numeric data 
# This df contains x variable and total_mobile_subscription variable
```

```python
# The first way to plot is to use the plt library
# Use pyplot library to plot out the data point x and its corresponding mobile subscription numbers.
# This makes a line plot, first input argument goes to the x-axis, second input argument goes to the y-axis.
plt.plot(df.x, df.total_mobile_subscription) # The line plot shows an overall upwards trend over times
```

```python
# Now, we try using data selection for plotting.
# The following code extracts the first 30 rows of data, and plot out the same
# line chart.
plt.plot(df.x[:30], df.total_mobile_subscription[:30]) 
```

```python
# This is the second way to plot, plot using dataframe
# The data must be available in the df.
# Here, it plots out the column of total_mobile_subscription using boxplot
# The following boxplot has a larger variation between median and upper 
# quartile as compared to between median and lower quarter

df.total_mobile_subscription.plot(kind="box")
```

```python
# Histogram can be plotted in a similar manner.
# The following code plots out the histogram of mobile subscription 
# with bin attribute set to be 10, i.e 10 bins will be created.
# The total range will be divided into 10 segments(bins) and its
# corresponding count/frequency is noted on the y-axis.
df.total_mobile_subscription.plot(kind="hist", bins=10)
```

```python
# The other useful plot is violinplot.
# This shows the min, max range like boxplot, but adds an additional
# dimension of distribution. The distribution is shown through the "fatness"
# of the curve colored in light blue. The "fat" area shows that a lot of data 
# is distributed in this area, like the value around 8000000.
# The "thinner" area shows lesser data distributed in this value,
# like around the value 5500000.
plt.violinplot(df.total_mobile_subscription, showmeans=True)
```

```python
# The third way to plot charts is to use subplots method.
# In this method, a function from plt "subplots" is called
# to get the figure and axes objects.
# Figure is the finished figure, ax / axes object is the object
# that does the crafting, and configuration of chart components
# The following line gets figure object and stored in a variable called fig
# axes object is stored in ax variable
fig, ax = plt.subplots() # get the figure and axes objects
# After getting axes object, set a title for the chart by calling the set_title
# on ax variable
ax.set_title("Mobile subscription")
# Next, call a boxplot to show on the same ax variable
# ax object can be used in the same manner to add more "features" to the chart.
ax.boxplot(df.total_mobile_subscription)
```

```python
# This cell shows another way to draw scatter plot using fig and ax methods
fig, ax = plt.subplots()
ax.scatter(df.x, df.total_mobile_subscription)
# The figure can be saved into a picture using the following line
# Call the savefig() function from fig variable, and provide the file name for chart
# The figure can be saved as png or jpg file.
# After running this command, you would be able to see a mobile_fig.png file
# generated in the same directory.
fig.savefig("mobile_fig.png")
```

```python
# When you have some simple data
# it can also be plotted using plotting library as below.
# making 2 simple lists, and draw a pie chart.
x = [50, 100, 80]
y = ["Cat1", "Cat2", "Cat3"]
plt.pie(x, labels=y)
```
